# Scraping Mentions
Scrapping all user one by one and gather all in a folder (`'mentions/'`). I save all of them in csv files. It may takes several hours. Among all non-API twitter `snscrape`, I think scraper is the fastest. but still, for this it takes hours. I take all mention tweet about those usernames in whole 2020

here is https://github.com/JustAnotherArchivist/snscrape a bit information about snscrape

```python
usernames = ['bankbri_id', 'kontakbri', 'promo_bri',
             'bankmandiri', 'mandiricare', 'wrausahamandiri', 'mandirifiesta', 'mandiricard',
             'kartukreditbca', 'bankbca', 'halobca', 'goodlifebca', 'xpresibca', 'bizguidebca',
             'bni', 'bniinstaplus', 'bnicreditcard', 'bnicardcenter']
```

```python
def scrapMentions(username):
    year = 2020
    months = range(1,13)
    df_all = pd.DataFrame()
    for month in months:
        last_day = str(calendar.monthrange(year, month)[1])
        month = str(month).zfill(2)
        a = '@{}'.format(username)
        b = "2020-{}-01".format(month)
        c = '2020-{}-{}'.format(month, last_day)
        cli_command = 'snscrape --jsonl --since {1} twitter-search "{0} until:{2}"> text-query-tweets.json'.format(a, b, c)
        os.system(cli_command)
        df_temp = pd.read_json("text-query-tweets.json", lines = True)
        df_all = df_all.append(df_temp)
        print("{}{}{} - {}".format(year, month, last_day, df_all.shape[0]))
    df_all.to_csv('mentions/{}.csv'.format(username))    
```

```python
for username in usernames:
    scrapMentions(username)
```

# Cleansing Tweets

```python
file_dir = 'mentions'
isfirst = True

for file in os.listdir(file_dir):
    username = file[:-4]
    if username in ['bankbri_id', 'kontakbri', 'promo_bri']:
        bank = 'BRI'
    elif username in ['bankmandiri', 'mandiricare', 'wrausahamandiri', 'mandirifiesta', 'mandiricard']:
        bank = 'Mandiri'
    elif username in ['kartukreditbca', 'bankbca', 'halobca', 'goodlifebca', 'xpresibca', 'bizguidebca']:
        bank = 'BCA'
    elif username in ['bni', 'bniinstaplus', 'bnicreditcard', 'bnicardcenter']:
        bank = 'BNI'
    print(username, bank)
    df_temp = pd.read_csv(file_dir + '/' + file)\
        .loc[:, ['id', 'date', 'content', 'user', 'replyCount', 'likeCount', 'quoteCount', 'lang', 'sourceLabel', 'mentionedUsers']]
    df_temp = df_temp[df_temp.lang == 'in']
    df_temp['user_username'] = df_temp['user'].apply(lambda x: eval(x)['username'])
    df_temp['user_verified'] = df_temp['user'].apply(lambda x: eval(x)['verified'])
    df_temp['user_created'] = df_temp['user'].apply(lambda x: eval(x)['created'])
    df_temp['user_followersCount'] = df_temp['user'].apply(lambda x: eval(x)['followersCount'])
    df_temp['user_friendsCount'] = df_temp['user'].apply(lambda x: eval(x)['friendsCount'])
    df_temp['user_statusesCount'] = df_temp['user'].apply(lambda x: eval(x)['statusesCount'])

    def listOfAccount(x):
        if str(x)!= 'nan':
            if [acc['username'] for acc in eval(str(x)) if acc['username'].lower() != username] != []:
                return [acc['username'] for acc in eval(str(x)) if acc['username'].lower() != username]
        else:
            return None
    df_temp['mentionedUsers_account'] = df_temp['mentionedUsers'].apply(listOfAccount)
    df_temp['mentionedUsers_count'] = df_temp['mentionedUsers_account'].apply(lambda x: len(x) if x != None else 0)

    df_temp = df_temp[df_temp.user_username.str.lower() != username]\
        .drop(columns = ['lang', 'user', 'mentionedUsers'])
    df_temp['account'] = username
    df_temp['bank'] = bank
    if isfirst:
        df_all = df_temp
        isfirst = False
    else:
        df_all = pd.concat([df_all, df_temp])
df_all.shape
```
